# Tarawera 100M 2019

Results below are from the Tarawera 100 2019 race. DNFs and DNSs are not included.

```r
library("readxl")

# xlsx files
tarawera <- read_excel("tarawera100m2019.xlsx")
```

```r
print(tarawera)
```

```r
library(ggplot2)
```

```r
ggplot(tarawera,aes(x=Time*24,group=Gender,fill=Gender))+
  geom_histogram(position="identity",alpha=0.5,binwidth=2)+theme_bw()
```

```r
ggplot(tarawera,aes(x=Time*24,group=Gender,fill=Gender)) + geom_density(alpha = 0.5)
```

```r
ggplot(tarawera,aes(x=Gender, y=Time*24, fill=Gender)) + geom_boxplot() + geom_jitter(position=position_jitter(0.1), alpha = 0.25)
```

### Average finish times by gender:

```r
f_mean = mean(tarawera$Time[tarawera$Gender == "F"])

print("Mean for females:")
print(f_mean * 24)

m_mean = mean(tarawera$Time[tarawera$Gender == "M"])

print("Mean for males:")
print(m_mean * 24)
```

# TAHOE 200M 2018

```r
ggplot(tahoe,aes(x=Time*24,group=Gender,fill=Gender))+
  geom_histogram(position="identity",alpha=0.5,binwidth=3)+theme_bw()
```

```r
ggplot(tahoe,aes(x=Time*24,group=Gender,fill=Gender)) + geom_density(alpha = 0.5)
```
